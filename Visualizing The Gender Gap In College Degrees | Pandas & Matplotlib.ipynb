# Visualizing The Gender Gap In College Degrees | Pandas & Matplotlib
---
In this project, I'm going to work with a dataset that contains information about the percentage of bachelor's degrees granted to women from 1970 to 2012 in the USA. 

Original data were released by [The Department of Education Statistics](http://nces.ed.gov/programs/digest/2013menu_tables.asp), but the dataset that I'm gonna work with had been already cleaned by Randal Olson, a data scientist at University of Pennsylvania. You can download it [here](http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv)
Randal compiled this data set to explore the gender gap in STEM fields, which stands for science, technology, engineering, and mathematics.

I'm gonna use `matplotlib` library for data visualization and apply Tableau colorblind palette for those colorblind readers.

# Exploring data
---

```python
import pandas as pd
import matplotlib.pyplot as plt
# plt.style.use('seaborn')
```

```python
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
```

# Visualizing data
---

```python
# using Tableu Color Blind 10 pallete for graphs:

cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show();
```

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

```python
fig = plt.figure(figsize=(15, 25))

# plotting STEM categories

for sp in range(0, 18, 3):
    cat_ind = int(sp/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[cat_ind]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[cat_ind]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[cat_ind])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    
# plotting Liberal arts categories

for sp in range(0, 15, 3):
    cat_ind = int(sp/3)
    ax = fig.add_subplot(6,3,sp+2)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[cat_ind]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[cat_ind]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[cat_ind])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    
# plotting Other categories:

for sp in range(0, 18, 3):
    cat_ind = int(sp/3)
    ax = fig.add_subplot(6,3,sp+3)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[cat_ind]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[cat_ind]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[cat_ind])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    if sp+3 == 3:
        plt.legend(loc='center right')
```
